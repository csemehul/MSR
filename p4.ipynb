```python
import pandas as pd
import matplotlib.pyplot as plt
# Read data from Excel
data = pd.read_excel('/content/drive/MyDrive/p4.xlsx')

# Select columns for your chart (replace 'column1' and 'column2' with actual names)
x_data = data['Column A (X-axis)']
y_data = data['Column B (Y-axis)']

# Create your chart (replace 'plot type' with your desired chart, e.g., 'bar', 'line', 'scatter')
plt.bar(x_data, y_data)

# Customize your chart (optional)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Chart Title')

# Save the chart as an image
plt.savefig('your_chart.png')  # Replace with desired format (e.g., 'png', 'jpg')

# Display the chart (optional)
plt.show()
```
